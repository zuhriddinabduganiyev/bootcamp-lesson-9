```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
playmarket_df = pd.read_csv("Google-Playstore.csv")
```

```python
playmarket_df.head(3)
```

### 1.PlayMarketdagi pulli va tekin ilovalar soni ?

```python
sns.countplot(x=playmarket_df["Free"], hue="Free", data=playmarket_df)
plt.figure(figsize=(12,6))
plt.show()
```

### 1. 2.5millionga yaqin ilovalar tekin va 140mingga yaqin ilovalar pulli ekanini ko'rishimiz mumkin. 

## 2. Ilovalarni cheklovlar soni (ya'ni kimlar install qilishi mumkin) ?

```python
plt.figure(figsize=[16,6])
sns.countplot(x=playmarket_df["Content Rating"], data=playmarket_df, hue=playmarket_df["Content Rating"], orient="x", saturation=2)
plt.xticks(rotation=45)
plt.title("Applarni cheklovlar soni (PlayMarketda)")
plt.show()
```

### 2. PlayMarketda 2million ilovalarni hamma yuklab olishi mumkin ekan, nechidir mingta ilovalarni faqatgina 10yoshdan yuqorilar yuklab olishi mumkin ekan.

### 4. Ilovalarda xaridlar soni kam ekan 250minga ham yetmayapti

```python
plt.figure(figsize=(14, 7))
plt.xticks(rotation=90)
plt.xlabel("Content Rating")
plt.ylabel("Maximum Installs")
plt.title("Content Rating vs Maximum Installs")
sns.scatterplot(data=playmarket_df, x="Maximum Installs", y="Rating Count", hue="Content Rating")
plt.show()
```

```python
top_category = playmarket_df.Category.value_counts().reset_index()
top_10_categories = top_category.head(10)
categories = top_10_categories['Category']
counts = top_10_categories['count']
plt.figure(figsize=(10, 6))
sns.barplot(x=categories, y=counts)
plt.title('Top 10 Categories')
plt.xlabel('Categories') 
plt.ylabel('Count')
plt.xticks(rotation=90)  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_category['Category'], y=top_category['count'])
plt.title('Top 10 Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
top_category_rating = playmarket_df.groupby('Category')['Rating'].mean().reset_index()
top_category_rating = top_category_rating.sort_values(by='Rating', ascending=False).head(10)
top_category_rating = playmarket_df.groupby('Category')['Rating'].mean().reset_index()
top_category_rating = top_category_rating.sort_values(by='Rating', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_category_rating['Category'], y=top_category_rating['Rating'], color="yellow")
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Top 10 Category with highest rating')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Highest Free and paid app category based on the Maximum installs')
sns.barplot(x='Category', y='Maximum Installs', data=playmarket_df, color="red")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Highest Free and paid app category based on the Maximum installs')
sns.barplot(x='Category', y='Maximum Installs', data=playmarket_df, hue='Free')
plt.show()
```
